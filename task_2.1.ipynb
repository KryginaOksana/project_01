```python
# Задача 2.1. 

# Создайте две функции maximum и minimum,
# которые получают список целых чисел в качестве входных данных 
# и возвращают наибольшее и наименьшее число в этом списке соответственно.


import random
lst = []
for _ in range(10):
    lst.append(random.randint(-1000, 1000))


def minimum(arr):
    x = arr[0]
    j = 0
    for _ in arr:
        if arr[j] < x:
            x = arr[j]
        j += 1
    return x

        
def maximum(arr):
    x = arr[0]
    j = 0
    for _ in arr:
        if arr[j] > x:
            x = arr[j]
        j += 1
    return x


print(f'Список {lst}, минимум: {minimum(lst)}, максимум: {maximum(lst)}')
   






```
